```python
import matplotlib.pyplot as plt

def plot_errorbars_only(df, node):
    # extract relevant columns
    untreated_col = 'Untreated-' + node
    diff_cols = [col for col in df.columns if col.startswith('Diff-') and col.endswith(node)]
    
    # get the min and max values for each row
    row_mins = df[diff_cols].min(axis=1)
    row_maxes = df[diff_cols].max(axis=1)
    
    # plot the error bars only
    fig, ax = plt.subplots()
    ax.errorbar(df.index, df[untreated_col], yerr=[row_mins, row_maxes], fmt='none', ecolor='b')
    ax.set_xlabel('Time')
    ax.set_ylabel(node)
    ax.set_title('Error bars of ' + node)
    plt.show()

for node in nodes:
    plot_errorbars_only(all_res,node)
```

```python
import matplotlib.pyplot as plt

def plot_with_errorbars(df, node):
    # extract relevant columns
    untreated_col = 'Untreated-' + node
    diff_cols = [col for col in df.columns if col.startswith('Diff-') and col.endswith(node)]
    
    # get the min and max values for each row
    row_mins = df[diff_cols].min(axis=1)
    row_maxes = df[diff_cols].max(axis=1)
    
    # plot the mean of min and max values as a horizontal line
    mean = (row_mins + row_maxes) / 2
    plt.axhline(mean.mean(), linestyle='--', color='gray')
    
    # plot the error bars
    fig, ax = plt.subplots()
    ax.errorbar(df.index, df[untreated_col], yerr=[row_mins, row_maxes], fmt='o')
    ax.set_xlabel('Time')
    ax.set_ylabel(node)
    ax.set_title('Untreated ' + node + ' with error bars')
    plt.show()

plot_with_errorbars(all_res,"M1")
```

```python
import matplotlib.pyplot as plt

def plot_with_error_bars(df, node):
    untreated_col = 'Untreated-' + node
    untreated_data = all_res[untreated_col]
    plt.plot(untreated_data.index, untreated_data.values, label=untreated_col)

    max_diff = -float('inf')
    min_diff = float('inf')
    for prefix in prefixes:
        for col in all_res.columns:
            if col.endswith('-' + node):
                if col.startswith('Diff-'):
                    max_diff = max(max_diff, all_res[col].max())
                    min_diff = min(min_diff, all_res[col].min())

    # Add error bars to untreated trajectory plot
    plt.fill_between(untreated_data.index,
                     untreated_data - min_diff,
                     untreated_data + max_diff,
                     alpha=0.2)
    
    plt.legend()
    plt.title(node)
    plt.show()

plot_with_error_bars(all_res,"M1")
```
