```python
df = pd.concat(map(f, glob.glob(os.path.join('', file_pattern))), axis= 1 )
df = df[df.columns.drop(list(df.filter(regex='Unnamed')))]
df['Time'] = df.index
```

```python
param = [os.path.basename(f) for f in glob.glob(file_pattern)]
param = [s.replace(name_pattern, '') for s in param]
param[1:5]
```

```python
all_res = pd.merge(df, Untreated, on='Time')
```

```python
prefixes = param
nodes = ['Prol_KC','preDiff_KC', 'Diff_KC', 'preDiff_KC', 'pDC', 'iDC', 'M1', 'M2', 'Neutrophil', 'Th0', 'Th1', 'Th17', 'Th22', 'Treg', 'Fibroblast']
    
for prefix in prefixes:
    for node in nodes:
        untreated_col = 'Untreated-' + node
        for col in all_res.columns:
            if col.endswith(prefix + '-' + node):
                diff_col_name = 'Dev-' + prefix + '-' + node
                all_res[diff_col_name] = all_res[untreated_col] - all_res[col]
```
